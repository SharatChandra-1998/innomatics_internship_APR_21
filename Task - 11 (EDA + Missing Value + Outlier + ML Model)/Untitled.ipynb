### BOX PLOTS :
A box-plot is a very useful and standardized way of displaying the distribution of data based on a five-number summary (minimum, first quartile, second quartile(median), third quartile, maximum). It helps in understanding these parameters of the distribution of data and is extremely helpful in detecting outliers.

```python
plt.boxplot(df['age'])
```

```python
#Draw a vertical boxplot grouped by a categorical variable
ax = sns.boxplot(x="gender", y="age", data=df)
```

```python
#Draw a boxplot with nested grouping by two categorical variables:
ax = sns.boxplot(x="gender", y="hours-per-week", hue="race",
                 data=df, palette="Set3")
```

```python
#Draw a boxplot with nested grouping when some bins are empty:
ax = sns.boxplot(x="income", y="age", hue="gender",
                 data=df, linewidth=2.5)
```

# Z-SCORE METHOD

Using Z score method,we can find out how many standard deviations value away from the mean  


```python
out=[]
def Zscore_outlier(df):
    m = np.mean(df)
    sd = np.std(df)
    for i in df: 
        z = (i-m)/sd
        if np.abs(z) > 3: 
            out.append(i)
    print("Outliers:",out)
Zscore_outlier(df['age'])
```
